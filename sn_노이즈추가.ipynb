```python
import subprocess

# 예제 입력과 출력 파일 지정
noisy_speech_file = "/drive/MyDrive/noisy_speech.raw"
denoised_output_file = "/drive/MyDrive/denoised_output.raw"

# rnnoise_demo 실행
subprocess.run(["./examples/rnnoise_demo", noisy_speech_file, denoised_output_file])

import wave
import numpy as np

# Raw PCM 파일 경로
raw_file_path = "/content/denoised_output.raw"


# WAV 파일 경로
wav_file_path = "/content/denoised_output.wav"


# Raw PCM 데이터 읽기
pcm_data = np.fromfile(raw_file_path, dtype=np.int16)

# WAV 파일로 쓰기
with wave.open(wav_file_path, 'w') as wav_file:
    wav_file.setnchannels(1)  # 1채널 (모노)
    wav_file.setsampwidth(2)   # 16-bit
    wav_file.setframerate(48000)  # 샘플링 레이트
    wav_file.writeframes(pcm_data.tobytes())
```
